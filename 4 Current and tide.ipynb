# Current and Tide Conditions

## 4.1 Current conditions
### 4.1.1 General

#### 4.1.1.1 
The effects of currents should be considered in design, construction and operation.

#### 4.1.1.2 
Items should be considered

* large steady excursions and slow drift motions of moored platforms
* rise to drag and lift forces on submerged structures
* Vortex induced vibrations/motions
* Interaction between strong currents and waves leads to change in H and T
* Scouring 

#### 4.1.1.3 
Current measurement campaigns recommended; should extend over the water column; capture several major storm events; Reference: ISO 19901-1(2005) _Metocean Design and operating considerations_.

#### 4.1.1.4 

Joint current-wave distribution of parameters and correspoding contour curves (surfaces). Or conservative method should be applied.(NORSOK N-003, OS-C101)




### 4.1.2 Types of current
#### 4.1.2.1 
Type:
* Wind generated currents
* tidal currents
* circulational currents
* loop and eddy currents
* soliton currents
* longshore currents

#### 4.1.2.2

Wind-generated currents: by wind stress and atmospheric pressure gradient 

#### 4.1.2.3

Tidal, HAT(Highest Astronomical Tide), LAT(Lowest ...)

#### 4.1.2.4

Oceanic ciculation: The [Gulf Stream](https://en.wikipedia.org/wiki/Gulf_Stream) in the Atlantic ocean. [Kuroshio Current](https://en.wikipedia.org/wiki/Kuroshio_Current) in the Pacific Ocean.


#### 4.1.2.5

Soliton currents are due to [Internal waves](https://en.wikipedia.org/wiki/Internal_wave).

#### 4.1.2.6

Loop/eddy currents and soliton currents penetrate deeply

#### 4.1.2.7

Longshore current: wave breaking(aka littoral current). [Radiation stress](https://en.wikipedia.org/wiki/Radiation_stress)

#### 4.1.2.8

Earthquakes can cause tubidity currents. Desing of pipelines. Tsunami


### 4.1.3 Current Velocity

#### 4.1.3.1

Varies with water depth, stretched or compressed due to surface waves

$v_{c}=v_{c}(x, y, x ; t)$

#### 4.1.3.2

For most cases: a steady flow field ; certain dynamic current sensitive structures: incuding turbulence effects.

#### 4.1.3.3

$v_{c}(z)=v_{c, \text { wind }}(z)+v_{c, t i d e}(z)+v_{c, \operatorname{circ}}(z)+\ldots$


### 4.1.4 Design current profiles
#### 4.1.4.1
Current conditions  are characterized by a current profile 

estimating current velocity 

corresponding to a given annual exceedance probability for each water depth

#### 4.1.4.2

Shallow tidal current velocity: power law, uni-directional current

$v_{c, t i d e}(z)=v_{c, t i d e}(0)\left(\frac{d+z}{d}\right)^{\alpha}$ for $z \leq 0$

#### 4.1.4.3 

The wind generated current can be taken as a linear profile from $z = -d_0$ to still water level

$v_{c, \text { wind }}(z)=v_{c, \text { wind }}(0)\left(\frac{d_{0}+z}{d_{0}}\right)$ for $-d_{0} \leq z \leq 0$

Or as a slab profle

$v_{c, \text { wind }}(z)=v_{c, \text { wind }}(0)$ for $-d_{0}<z<0$


#### 4.1.4.4

Wind generated current is limited in depth ($d_0 < 50$m)

#### 4.1.4.5

In deep water along an open coastline, wind-generated current velocity at still water level

$v_{c, \text { wind }}(0)=k U_{1 \text { hour, } 10 m}$ where $k=0.015$ to $0.03$

#### 4.1.4.6

The variation of current velocity over depth may be complex.

#### 4.1.4.7

When long-term measured current profile dataset are available, Emprical Orthogonal Functions(EOF) can be applied.

$v_{c}(x, t)=\sum_{m=1}^{M} b_{m}(x) \cdot w_{m}(t)$

#### 4.1.4.8 
Profiles with required return periods can be selected by EOFs and IFORM method.

#### 4.1.4.9

Current spectra are not in general available for current loading except for turbulent boundary layer close to the sea bed. refer to RP-F105
