## 3.6 Long term wave statistics

### 3.6.1 Analysis strategies

#### 3.6.1.1
The long-term variation of wave climate can be described as generic distributions or scatter diagrams for governing sea state parameters, such as $(H_s, T_z, \theta)$

#### 3.6.1.2

Scatter diagram, marginal distribution and joint environmental models

#### 3.6.1.3

Two analysis strategies :
* The global model(initial distribution method): all available data
* The event model: Observations over some threshold(POT or strom analysis method, annual extremes or seasonal extremes)

#### 3.6.1.4
The initial distribution method  is typically applied for the distribution of sea state parameters such as $H_s$
The event based approaches can be applied for sea state parameters, but might also be used directly for ... <!---maximum individual wave height and for the maximum crest height.-->

#### 3.6.1.5
Trade-off between two methods

#### 3.6.1.6
Fitting techniques:

* method of moments (MOM)
* probability weighted moments (PWM)
* least squares methods (LS) 
* maximum likelihood estimation (MLE)

#### 3.6.1.7
* MOM: better at the mode of the distribution
* MLE: theoretical advantages, but can be cumbersome in practice
* LS: typically more influenced by the tail behaviour than are MOM and MLE.

#### 3.6.1.8
When estimating extremes, it is important that the tail of the fitted distribution honour the data and for 3-parameter Weibull distribution the LS fit often gives better tail fit than the method of moments. 

For other applications, a good fit to the main bulk of data might be more important.

### 3.6.2 Marginal distribution of siginicant wave height

#### 3.6.2.1

$F_{H_{S}}(h)=1-\exp \left[-\left(\frac{h-\gamma}{\alpha}\right)^{\beta}\right]$

where $\alpha$ is the scale parameter, $\beta$ is the shape parameter, and $\gamma$ is the location parameter (lower threshold)
#### 3.6.2.2
For  POT and storm statistics analysis, a 2-parameter weibull distribution or an exponential distribution is recommended

$F_{E}(h)=\exp \left\{-\frac{h-h_{0}}{\theta}\right\}$

#### 3.6.2.3
Sensitivity of adopted threshold level should be performed, as well as comparison of alternative method results.

#### 3.6.2.4

#### 3.6.3.5
CMA joint model for significant wave height and zero up-crossing period is recommended:

$f_{H_{s} T_{z}}(h, t)=f_{H_{S}}(h) \cdot f_{T_{z} \mid H_{s}}(t \mid h)$

the significant wave height is modelled by a 3-parameter Weibull probability density function

$f_{H_{S}}(h)=\frac{\beta_{H s}}{\alpha_{H s}}\left(\frac{h-\gamma_{H S}}{\alpha_{H s}}\right)^{\beta_{H s}-1} \exp \left\{-\left(\frac{h-\gamma_{H s}}{\alpha_{H s}}\right)^{\beta_{H s}}\right\}$

zero-crossing wave period conditional on Hs is modelled by a lognormal distribution

$f_{T_{Z} \mid H_{S}}(t \mid h)=\frac{1}{\sigma t \sqrt{2 \pi}} \exp \left\{\frac{(\ln t-\mu)^{2}}{2 \sigma^{2}}\right\}$

distribution parameters $\mu$ and $\sigma$ are functions of the significant wave height 

$\mu=E\left[\ln T_{z}\right]=a_{0}+a_{1} h^{a_{2}}$

$\sigma=s t d\left[\ln T_{z}\right]=b_{0}+b_{1} e^{b_{2} h}$

### 3.6.4 Joint distribution of significan wave height and wind speed

$f_{U \mid H_{S}}(u \mid h)=k \frac{u^{k-1}}{U_{c}^{k}} \exp \left[-\left(\frac{u}{U_{c}}\right)^{k}\right]$

### 3.6.7 Long term distribution of individual wave height

The long term distribution $F_H(x)$  of individual wave height can be obtained by integrating the short term distribution $F_{H \mid H_{S}}\left(x \mid h_{s}\right)$

$F_{H}(x)=\frac{1}{\bar{v}} \int_{h} \int_{t} T_{z}^{-1} \cdot F_{H \mid H_{S}}(x \mid h) f_{H_{s} T_{Z}}(h, t) d h d t$

$=\frac{1}{\bar{v}_{0}} \int_{h}\left|T_{z}^{-1}\right|_{h}{F}_{H \mid H_{s}}(x \mid h) f_{H_{s}}(h) d h$

where

$\left\langle T_{z}^{-1}\right\rangle_{h}=\int_{t} T_{z}^{-1} f_{\left.T_{z}\right|_{H_{s}}}(t \mid h) d t$

$\bar{v}_{0}=\int_{h} \int_{t} T_{z}^{-1} f_{H_{s} T_{z}}(h, t) d h d t$

$=\int_{h}\left\langle T_{z}^{-1}\right\rangle_{h} f_{H_{s}}(h) d h$

The individual wave height $H_{T_{R}}$   with return period $T_{R}$ (in years) follows from:

$1-F_{H}\left(H_{T_{R}}\right)=\frac{1}{{T}_{R} \cdot 365.25 \cdot 24 \cdot 3600 \cdot \bar{v}_{0}}$



## 3.7 Extreme value distribution

### 3.7.1 Design sea state
#### 3.7.1.1 
$F_{H_{s}}(h)$ denotes the distribution of the significant wave height in an arbitrary t-hour sea state,the distribution of the annual maximum significant wave height Hs,max can be taken as

$F_{H_{s, \mathrm{max}^{\prime}}}$ 1year $(h)=\left(F_{\mathrm{H}_{s}}(h)\right)^{n}$

where $n$ is the number of t-hour sea states in one year. For $t = 3$ hours, $n = 2922$

#### 3.7.1.2

$H_{s, T_{R}}=F_{H_{S}}\left(1-\frac{1}{n T_{R}}\right)$

#### 3.7.1.3

Alternatively, $H_{S,T_{R}}$can be defined as the $(1-1/T_R)$ quantile in the distribution of the annual maximum significant wave height

$H_{s, T_{R}}=F_{H_{s, \max ^{\prime} 1year }}$ $^{-1}\left(1-\frac{1}{T_{R}}\right) ; T_{R}>1 year$ 

#### 3.7.1.4

The short term variability of response within the sea state should be taken into consideration, confer [3.7.2].

### 3.7.2 Metocean contours
#### 3.7.2.1
Definition
#### 3.7.2.2
IFORM approach

- Determine the joint model of sea state variables of interest:

For $H_{s}, T_{z}:  f_{H_{s} T_{z}}(h, t)=f_{H_{s}}(h) f_{\left.T_{z}\right|H_{s}}(t \mid h)$
- Transform the distribution to standard normalized U-space:

$\Phi\left(u_{1}\right)=F_{H_{s}}\left(h_{s}\right) \quad \Phi\left(u_{2}\right)=F_{T_{z} \mid H_{s}}\left(t_{z}\right)$

- Establish the circle for prescribed return period in U-space

$\sqrt{u_{1}^{2}+u_{2}^{2}}=\beta=-\Phi^{-1}\left[\frac{1}{100 \cdot 365 \cdot 8}\right]=4.5$

- ransform the circle into a contour in the sea state parameter space

$h_{s}=F_{H_{s}}^{-1}\left(\Phi\left(u_{1}\right)\right) \quad t_{z}=F_{T_{z} \mid H_{s}}^{-1}\left(\Phi\left(u_{2}\right)\right)$

#### 3.7.2.3
An alternative approach to metocean contours is based on direct Monte Carlo simulations from the joint
model of sea state variables of interest

- Determine the joint model of sea state variables of interest $H_{s}, T_{z}:  f_{H_{s} T_{z}}(h, t)$

-  Simulate a sufficient number of Monte Carlo samples from the joint distribution model. Importance sampling may be used.

- For a suitable number of directions, $\theta$, calculate the points along the contour so that the ratio of samples falling outside a hyperplane defined by $\theta$ correspond to the desired return period.

#### 3.7.2.4

Constant probability density approach

- Determine the joint model of sea state variables of interest 

For $H_{s}, T_{z}:  f_{H_{s} T_{z}}(h, t)=f_{H_{s}}(h) f_{\left.T_{z}\right|H_{s}}(t \mid h)$

-  Estimate the extreme value for the governing variable for the prescribed return period, and associated values for other variables. For example 100-year value for $H_s$ and the conditional median for $T_z$

- The contour line is estimated from the joint model or scatter diagram as the contour of constant probability density going through the above mentioned parameter combination.


#### 3.7.2.5

An estimate of the extreme response is then obtained by searching along the metocean contour for the condition giving maximum characteristic extreme response.

#### 3.7.2.6

This method will tend to underestimate extreme response levels

#### 3.7.2.7

One can estimate the indirectly and approximately account for the extreme response variability by inflating the return period and metocean contours

#### 3.7.2.8

Inflate response: One can replace the stochastic response by a fixed fractile level higher than the median value, or apply multipliers of the median extreme response estimates

#### 3.7.2.9

The appropriate fractile levels and multipliers will be case-specific, and should be specified for classes of structures and structural responses

### 3.7.3 Extreme individual wave height and extreme crest height

#### 3.7.3.1 

The maximum individual wave height in a sea state

$F_{H_{\max }}\left(h_{\max }\right)=$
$\left.\int_{h_{s}} \int_{t_{z}} F_{H_{\max }}\right|_{s_{s}^{T} z}\left(h_{\max } \mid h_{s^{\prime}} t_{z}\right) f_{H_{s} T_{z}}\left(h_{s^{\prime}} t_{z}\right) d h_{s} d t_{z}$

Here

$f_{H_{s} T_{z}}\left(h_{s^{\prime}} t_{z}\right)$ is the joint probability density for $H_s$ and $T_z$

$F_{H_{\max }} \mid H_{s^{T} z}\left(h_{\max } \mid h, t\right)$


is the distribution of the maximum wave height in the sea state with parameters $H_s$,$T_z$. A corresponding expression applies for crest height and for storm events.


#### 3.7.3.2

The following recipe is recommended to establish the distribution for the extreme waves based on storm statistics:

- step through storms, establish distribution of maximum wave height in the storm as Gumbel distribution to $h_{max}^2$

$ F_{H_{\max}}(h_{\max})\approx \exp \left\{-\exp \left(-\ln (N) \cdot\left[\left(\frac{h_{\max }}{h_{\max , m}}\right)^{2}-1\right]\right)\right\}$

- determine mode $h_{max,m}$ and variance for the extreme value distribution within each storm

- carry out POT analysis for the modes $h_{max,m}$

The parameter $ln(N)$ is a function of the coefficient of variation of $h_{max}^2$


#### 3.7.3.3
The annual extreme value distributions for wave height are obtained by integrating the short term statistics weighted by the long term distributions

$F_{H_{\text {annual }}}\left(h_{\max }\right)=\left[F_{H \max }\left(h_{\max }\right)\right]^{n}$

#### 3.7.3.4
Assuming a sea state duration 3-hours, the value $H_{T_R}$ with return period $T_R$ (in years) follows from:

$1-F_{H_{\max }}\left(H_{T_{R}}\right)=\frac{1}{T_{R} \cdot 2922}$

$1-F_{H_{\text {annual }}}\left(H_{T_{R}}\right)=\frac{1}{T_{R}}$

#### 3.7.3.5

In lack of more detailed information, for sea states of duration 3-hours, the $H_100$ may be taken as 1.9 times the significant wave height $H_{s,100}$.
